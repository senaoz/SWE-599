```python
import json
import pandas as pd
import ast
import numpy as np

# Load data
boun_data = pd.read_csv("data/cleaned/boun.csv")
priority_followed_data = pd.read_csv("data/cleaned/priority_followed.csv")

def safe_parse(x):
    """Safely parse stringified JSON/lists using ast.literal_eval with fallback."""
    if pd.isna(x):
        return None
    
    if isinstance(x, str):
        x = x.strip()
        if x.startswith(('[{', '[', '{', '(')):  # Detect literal structures
            try:
                return ast.literal_eval(x)
            except (ValueError, SyntaxError):
                # Fallback to string or attempt json.loads for valid JSON
                try:
                    return json.loads(x)
                except:
                    return x
    return x

# Apply parsing with progress tracking
print("Parsing BOUN data...")
for col in boun_data.columns:
    print(f"  Processing {col}...")
    boun_data[col] = boun_data[col].map(safe_parse)

print("Parsing priority_followed data...")
for col in priority_followed_data.columns:
    print(f"  Processing {col}...")
    priority_followed_data[col] = priority_followed_data[col].map(safe_parse)
```

```python
import pandas as pd
import json
import os

def create_benchmark_pairs(priority_df, boun_df, output_path='eval_dataset/benchmark_pairs.json'):
    # Klasör kontrolü
    output_dir = os.path.dirname(output_path)
    if output_dir and not os.path.exists(output_dir):
        os.makedirs(output_dir)

    benchmark_data = []
    
    # Veriyi temizle (Abstract olmayanlarla embedding yapamazsın)
    priority_df = priority_df.dropna(subset=['abstract']).copy()
    boun_df = boun_df.dropna(subset=['abstract']).copy()

    # Örnekleme: Tüm dataset yerine test için ilk 100 makaleyi sorgu yapalım
    test_queries = priority_df.head(100)
    
    for _, query_row in test_queries.iterrows():
        # OpenAlex'in 'related_works' listesini kontrol et
        related_ids = query_row.get('related_works', [])
        if not isinstance(related_ids, list): related_ids = []
        
        # 1. Direct Match: Related works listesinde BOUN makalesi var mı?
        positives = boun_df[boun_df['id'].isin(related_ids)].copy()
        
        # 2. Concept Match: Eğer direct match yoksa aynı konsepttekileri bul
        if positives.empty:
            q_concepts = [c['id'] for c in query_row['concepts']] if isinstance(query_row['concepts'], list) else []
            if q_concepts:
                positives = boun_df[boun_df['concepts'].apply(
                    lambda x: any(c['id'] in q_concepts for c in x) if isinstance(x, list) else False
                )].head(3) # En güçlü 3 tanesini al

        if not positives.empty:
            benchmark_data.append({
                "query": {
                    "id": query_row['id'],
                    "title": query_row['title'],
                    "abstract": query_row['abstract']
                },
                "ground_truth_positives": positives[['id', 'title']].to_dict('records')
            })

    # Dosyaya yaz
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(benchmark_data, f, ensure_ascii=False, indent=4)
    
    return len(benchmark_data)
```

```python
pairs_count = create_benchmark_pairs(priority_followed, boun)
pairs_count
```

```python
import json
import pandas as pd

# 1. Dosyayı oku
with open('eval_dataset/benchmark_pairs.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

flat_list = []
for entry in data:
    query_title = entry['query']['title']
    query_abstract = entry['query']['abstract']
    # Her sorgu için eşleşen ilk BOUN makalesini alalım (Örnekleme için)
    if entry['ground_truth_positives']:
        match = entry['ground_truth_positives'][0]
        flat_list.append({
            "Global_Query_Text": query_title + " " + query_abstract, 
            "BOUN_Match_Text": match['title'],
            "Match_Count": len(entry['ground_truth_positives']) # Bu sorgu için toplam kaç BOUN eşleşmesi var?
        })

df_view = pd.DataFrame(flat_list)

pd.set_option('display.max_colwidth', None)

# sort by Match_Count
df_view = df_view.sort_values(by='Match_Count', ascending=False)
df_view
```
