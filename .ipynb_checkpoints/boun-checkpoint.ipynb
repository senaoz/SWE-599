```python
import pandas as pd

from openalex_client import BOUN_OPENALEX_ID, FOLLOWED_INSTITUTIONS

import requests
import pandas as pd
import time
from urllib.parse import quote

def fetch_all_boun_papers(output_file="data/boun_corpus.csv"):
    # Use canonical BOUN ID from openalex_client (Boğaziçi University)
    base_url = f"https://api.openalex.org/works?filter=authorships.institutions.lineage%3Ai4405392"
    
    all_papers = []
    page = 1
    
    print("Starting to fetch BOUN corpus...")
    
    while True:
        url = f"{base_url}&per_page=200&page={page}"
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"Error at page {page}: {response.status_code}")
            break
            
        data = response.json()
        results = data.get('results', [])
        
        if not results:
            break
            
        for work in results:
            all_papers.append({
                'id': work.get('id'),
                'title': work.get('display_name'),
                'publication_year': work.get('publication_year'),
                'type': work.get('type'),
                # Abstract - 'inverted index'
                'abstract_inverted': work.get('abstract_inverted_index'),
                'concepts': [c.get('display_name') for c in work.get('concepts', [])[:3]]
            })
            
        print(f"Page {page} fetched. Total papers: {len(all_papers)}")
        
        if len(results) < 200:
            break
        
        page += 1
        time.sleep(0.1)

    df = pd.DataFrame(all_papers)
    df.to_csv(output_file, index=False, encoding="utf-8")
    print(f"\nCompleted! Total {len(df)} papers saved to {output_file}")

if __name__ == "__main__":
    # fetch_all_boun_papers()
```
